## Creating Target and Feature data

### **Working with Model**

## Splitting data in train & test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Applying Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier()
model.fit(x_train,y_train)
```

## Train Score

```python
model.score(x_train,y_train)
```

## Test Score

```python
model.score(x_test,y_test)
```

## Y predict

```python
y_pred=model.predict(x_test)
```

## Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

## Trying for 1 to 14 values for k

```python
from sklearn.neighbors import KNeighborsClassifier

train=[]
test=[]
for i in range(1,15):
  model=KNeighborsClassifier(i)
  model.fit(x_train,y_train)
  train.append(model.score(x_train,y_train))
  test.append(model.score(x_test,y_test))
```

```python
test
```

```python
train
```

## Plotting Graph for K from 1 to 14

```python
import seaborn as sns
plt.figure(figsize=(5,5))
sns.lineplot(x=range(1,15),y=train,marker="o",label="Train_Score")
sns.lineplot(x=range(1,15),y=test,marker="o",label="Test_Score")
```
